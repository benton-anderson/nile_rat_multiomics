```r
library('lme4')
```

```r
data <- read.csv('..\\data\\interim\\data_for_lme4_R.csv')
data$animal  <- as.factor(data$animal)
data$week  <- as.factor(data$week)
data$sampling <- as.factor(data$sampling)
```

```r
data
```

```r
feature <- 'l_704'

full <- lmer(paste(feature, "~ week + sampling + ogtt + sampling:ogtt + (1|animal)"), data=data, REML=FALSE)
week <- lmer(paste(feature, "~        sampling + ogtt + sampling:ogtt + (1|animal)"), data=data, REML=FALSE)
samp <- lmer(paste(feature, "~ week +            ogtt + sampling:ogtt + (1|animal)"), data=data, REML=FALSE)
ogtt <- lmer(paste(feature, "~ week + sampling +        sampling:ogtt + (1|animal)"), data=data, REML=FALSE)
inte <- lmer(paste(feature, "~ week + sampling + ogtt +                 (1|animal)"), data=data, REML=FALSE)

anova(full, week, samp, ogtt, inte)
```
